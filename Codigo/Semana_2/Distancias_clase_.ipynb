## 1. Distancia Euclidiana
La distancia euclidiana es la distancia más común y se calcula como la longitud de la línea recta entre dos puntos en un espacio euclidiano.





image.png

Analogía: Imagina que estás en un plano cartesiano (como un mapa de ciudad) y necesitas encontrar la distancia más corta entre dos puntos. La distancia euclidiana sería la distancia en línea recta, como si pudieras volar entre los edificios en lugar de caminar por las calles.

Aplicación: Muy utilizada en algoritmos como K-Nearest Neighbors (KNN) y en clustering (como K-means). Es efectiva cuando las características están en la misma escala. (Es necesario que las variables esten escalas )

Por ejemplo, puede emplearse para agrupar puntos de datos similares o para encontrar los vecinos más cercanos de un punto de datos dado. ( pocas dimensiones )

```python
import numpy as np

# Definimos dos puntos en un espacio 2D
p = np.array([1, 2])
q = np.array([4, 6])

# Definimos una función para calcular la distancia de Minkowski
def minkowski_distance(p, q, lambda_val):
    return np.sum(np.abs(p - q) ** lambda_val) ** (1 / lambda_val) # lambda_val es la P

# Distancia de Manhattan (λ = 1)
dist_manhattan = minkowski_distance(p, q, 1)
print(f"Distancia de Manhattan: {dist_manhattan}")

# Distancia Euclidiana (λ = 2)
dist_euclidiana = minkowski_distance(p, q, 2)
print(f"Distancia Euclidiana: {dist_euclidiana}")

# Distancia de Chebyshev (λ = ∞, aproximado con un número muy grande)
dist_chebyshev = minkowski_distance(p, q, np.inf)
#dist_chebyshev = minkowski_distance(p, q, 2)
print(f"Distancia de Chebyshev: {dist_chebyshev}")


```

### Cálculo de la similaridad coseno

image.png

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    # Asegurate que ambos sean numpy arrays
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)

    # producto punto
    dot_product = np.dot(vec1, vec2)

    # Ccomputar normas o magnitudes
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)

    # Calculando la similaridad coseno
    cosine_sim = dot_product / (norm_vec1 * norm_vec2)

    return cosine_sim

# ejemplo de uso
vector1 = [1, 2, 3]
vector2 = [8, 10, 12]

similarity = cosine_similarity(vector1, vector2)
print(f"Cosine Similarity: {similarity}")
```

Esta distancia es útil en problemas de clasificación y detección de anomalías, ya que permite medir cuán diferente es un punto respecto a la distribución de datos, considerando las relaciones entre las variables.

```python
from scipy.spatial import distance # existe la libreria de mahalanobis

# Matriz de covarianza regularizada
cov_matrix = np.cov(np.array([p, q]).T) # traspuesta
reg_cov_matrix = cov_matrix + np.eye(cov_matrix.shape[0]) * 1e-5  # Añadir regularización a la diagonal , la diagonal de la matriz de covarianza

# Inversa de la matriz de covarianza regularizada
inv_cov_matrix = np.linalg.inv(reg_cov_matrix)

# Distancia de Mahalanobis
dist_mahalanobis = distance.mahalanobis(p, q, inv_cov_matrix)
print(f"Distancia de Mahalanobis: {dist_mahalanobis}")
```

```python
from sklearn.datasets import load_iris
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, cosine, mahalanobis #Cityblock == Manhattan
import pandas as pd

# Cargar el Iris dataset
iris = load_iris()
X = iris.data  # Características: Largo y ancho de sépalos y pétalos
y = iris.target  # Etiquetas de clase

# Crear un DataFrame para facilitar el manejo
df = pd.DataFrame(X, columns=iris.feature_names)

# Seleccionar dos muestras del dataset
sample1 = df.iloc[0].values
sample2 = df.iloc[50].values

# Calcular Distancias
dist_euclidiana = euclidean(sample1, sample2)
dist_manhattan = cityblock(sample1, sample2)
dist_coseno = cosine(sample1, sample2)

# Calcular la Distancia de Mahalanobis (requiere la matriz de covarianza)
cov_matrix = np.cov(X.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)
dist_mahalanobis = mahalanobis(sample1, sample2, inv_cov_matrix)

print(f"Distancia Euclidiana: {dist_euclidiana}")
print(f"Distancia de Manhattan: {dist_manhattan}")
print(f"Distancia Coseno: {dist_coseno}")
print(f"Distancia de Mahalanobis: {dist_mahalanobis}")
```

### 1. Distancia Euclidiana: 4.003748243833521
Interpretación:

La distancia euclidiana entre las dos muestras seleccionadas del Iris dataset es de aproximadamente 4.0. Esta métrica mide la distancia "en línea recta" o "a vuelo de pájaro" entre dos puntos en el espacio de características.
Una distancia euclidiana de 4.0 indica que los dos puntos están moderadamente separados en el espacio de características. Cuanto mayor sea esta distancia, más diferentes son los dos puntos (flores) en términos de sus características (longitud y ancho de sépalos y pétalos).
Contexto:

Una distancia baja podría significar que las flores son de especies similares o tienen características muy similares. Una distancia alta podría sugerir que pertenecen a especies diferentes o que las muestras están muy alejadas en términos de sus medidas.

### 2. Distancia de Manhattan: 6.7
Interpretación:

La distancia de Manhattan, también conocida como distancia L1, es la suma de las distancias absolutas entre las características correspondientes de las dos muestras. Aquí, la distancia de Manhattan es de 6.7.
Esto significa que si se midiera la distancia moviéndose solo a lo largo de las líneas rectas del eje (como moverse por una cuadrícula de calles), la distancia total entre las dos muestras sería 6.7.
Contexto:

La distancia de Manhattan tiende a ser más grande que la Euclidiana cuando se consideran varias dimensiones, ya que acumula las diferencias en cada dimensión sin elevarlas al cuadrado. Proporciona una visión alternativa de la diferencia, especialmente útil en datos con alta dimensionalidad o características escasas.

### 3. Distancia Coseno: 0.07161964128508791
Interpretación:

La distancia coseno se mide como 1 menos la similitud del coseno entre los dos vectores. Un valor de 0.0716 indica que el ángulo entre los dos vectores (muestras) es relativamente pequeño, lo que significa que los vectores apuntan en direcciones bastante similares.
En otras palabras, la orientación (pero no la magnitud) de las características de las dos flores es muy parecida.
Contexto:

La similitud del coseno es especialmente útil cuando queremos saber si dos muestras tienen una "dirección" similar en el espacio de características, independientemente de cuán largas o cortas sean las diferencias en sus características. Este valor bajo de distancia coseno indica que las muestras están orientadas en una dirección similar.

### 4. Distancia de Mahalanobis: 2.474107848855276
Interpretación:

La distancia de Mahalanobis es de aproximadamente 2.47, lo cual toma en cuenta la correlación entre las variables en el conjunto de datos. Esta métrica mide la distancia de una muestra al centroide de las muestras (media multivariante), normalizada por la dispersión de los datos.
En otras palabras, una distancia de Mahalanobis de 2.47 indica que las muestras están a 2.47 desviaciones estándar de diferencia cuando se tiene en cuenta la estructura de correlación del dataset.
Contexto:

Una distancia de Mahalanobis más baja sugiere que las muestras están más cerca del centroide en el contexto de la distribución de los datos, mientras que una distancia más alta indica que están más alejadas y pueden considerarse más "anómalas" o diferentes.

### Conclusión General:

Distancia Euclidiana y Manhattan: Ambas muestran que las dos muestras están moderadamente separadas en el espacio de características, pero la distancia Manhattan refleja una ruta más "realista" a lo largo de cada eje.

Distancia Coseno: Muestra que las dos muestras tienen una orientación similar, sugiriendo que, aunque pueden tener diferencias en la magnitud de algunas características, estas diferencias ocurren en la misma dirección.

Distancia de Mahalanobis: Considera la correlación entre características y muestra que las dos muestras no son extremadamente diferentes cuando se tienen en cuenta estas correlaciones.



```python
from sklearn.datasets import load_wine
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, cosine, mahalanobis #cityblock es manhathan
import pandas as pd

# Cargar el Wine dataset
wine = load_wine()
X = wine.data  # Características
y = wine.target  # Etiquetas de clase

# Crear un DataFrame para facilitar el manejo
df = pd.DataFrame(X, columns=wine.feature_names)

# Seleccionar dos muestras del dataset para comparar (por ejemplo, la primera y la segunda)
sample1 = df.iloc[0].values #Va al indice 0 y trae todos los datos
sample2 = df.iloc[100].values #Va al indice 10 y trae todos los datos

# Calcular Distancias
dist_euclidiana = euclidean(sample1, sample2)
dist_manhattan = cityblock(sample1, sample2)
dist_coseno = cosine(sample1, sample2)

# Calcular la Distancia de Mahalanobis (requiere la matriz de covarianza)
cov_matrix = np.cov(X.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)
dist_mahalanobis = mahalanobis(sample1, sample2, inv_cov_matrix)

# Imprimir resultados
print(f"Distancia Euclidiana: {dist_euclidiana}")
print(f"Distancia de Manhattan: {dist_manhattan}")
print(f"Distancia Coseno: {dist_coseno}")
print(f"Distancia de Mahalanobis: {dist_mahalanobis}")
```
