## 1. Distancia Euclidiana
La distancia euclidiana es la distancia ms comn y se calcula como la longitud de la lnea recta entre dos puntos en un espacio euclidiano.

Aplicacin: Muy utilizada en algoritmos como K-Nearest Neighbors (KNN) y en clustering (como K-means). Es efectiva cuando las caractersticas estn en la misma escala. (Es necesario que las variables esten escalas )

```python
import numpy as np

# Definimos dos puntos en un espacio 2D
p = np.array([1, 2])
q = np.array([4, 6])

# Definimos una funcin para calcular la distancia de Minkowski
def minkowski_distance(p, q, lambda_val):
    return np.sum(np.abs(p - q) ** lambda_val) ** (1 / lambda_val) # lambda_val es la P

# Distancia de Manhattan ( = 1)
dist_manhattan = minkowski_distance(p, q, 1)
print(f"Distancia de Manhattan: {dist_manhattan}")

# Distancia Euclidiana ( = 2)
dist_euclidiana = minkowski_distance(p, q, 2)
print(f"Distancia Euclidiana: {dist_euclidiana}")

# Distancia de Chebyshev ( = , aproximado con un nmero muy grande)
dist_chebyshev = minkowski_distance(p, q, np.inf)
#dist_chebyshev = minkowski_distance(p, q, 2)
print(f"Distancia de Chebyshev: {dist_chebyshev}")


```

### Clculo de la similaridad coseno

image.png

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    # Asegurate que ambos sean numpy arrays
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)

    # producto punto
    dot_product = np.dot(vec1, vec2)

    # Ccomputar normas o magnitudes
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)

    # Calculando la similaridad coseno
    cosine_sim = dot_product / (norm_vec1 * norm_vec2)

    return cosine_sim

# ejemplo de uso
vector1 = [1, 2, 3]
vector2 = [8, 10, 12]

similarity = cosine_similarity(vector1, vector2)
print(f"Cosine Similarity: {similarity}")
```

Esta distancia es til en problemas de clasificacin y deteccin de anomalas, ya que permite medir cun diferente es un punto respecto a la distribucin de datos, considerando las relaciones entre las variables.

```python
from scipy.spatial import distance # existe la libreria de mahalanobis

# Matriz de covarianza regularizada
cov_matrix = np.cov(np.array([p, q]).T) # traspuesta
reg_cov_matrix = cov_matrix + np.eye(cov_matrix.shape[0]) * 1e-5  # Aadir regularizacin a la diagonal , la diagonal de la matriz de covarianza

# Inversa de la matriz de covarianza regularizada
inv_cov_matrix = np.linalg.inv(reg_cov_matrix)

# Distancia de Mahalanobis
dist_mahalanobis = distance.mahalanobis(p, q, inv_cov_matrix)
print(f"Distancia de Mahalanobis: {dist_mahalanobis}")
```

```python
from sklearn.datasets import load_iris
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, cosine, mahalanobis #Cityblock == Manhattan
import pandas as pd

# Cargar el Iris dataset
iris = load_iris()
X = iris.data  # Caractersticas: Largo y ancho de spalos y ptalos
y = iris.target  # Etiquetas de clase

# Crear un DataFrame para facilitar el manejo
df = pd.DataFrame(X, columns=iris.feature_names)

# Seleccionar dos muestras del dataset
sample1 = df.iloc[0].values
sample2 = df.iloc[50].values

# Calcular Distancias
dist_euclidiana = euclidean(sample1, sample2)
dist_manhattan = cityblock(sample1, sample2)
dist_coseno = cosine(sample1, sample2)

# Calcular la Distancia de Mahalanobis (requiere la matriz de covarianza)
cov_matrix = np.cov(X.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)
dist_mahalanobis = mahalanobis(sample1, sample2, inv_cov_matrix)

print(f"Distancia Euclidiana: {dist_euclidiana}")
print(f"Distancia de Manhattan: {dist_manhattan}")
print(f"Distancia Coseno: {dist_coseno}")
print(f"Distancia de Mahalanobis: {dist_mahalanobis}")
```

### 1. Distancia Euclidiana: 4.003748243833521
Interpretacin:

La distancia euclidiana entre las dos muestras seleccionadas del Iris dataset es de aproximadamente 4.0. Esta mtrica mide la distancia "en lnea recta" o "a vuelo de pjaro" entre dos puntos en el espacio de caractersticas.
Una distancia euclidiana de 4.0 indica que los dos puntos estn moderadamente separados en el espacio de caractersticas. Cuanto mayor sea esta distancia, ms diferentes son los dos puntos (flores) en trminos de sus caractersticas (longitud y ancho de spalos y ptalos).
Contexto:

Una distancia baja podra significar que las flores son de especies similares o tienen caractersticas muy similares. Una distancia alta podra sugerir que pertenecen a especies diferentes o que las muestras estn muy alejadas en trminos de sus medidas.

### 2. Distancia de Manhattan: 6.7
Interpretacin:

La distancia de Manhattan, tambin conocida como distancia L1, es la suma de las distancias absolutas entre las caractersticas correspondientes de las dos muestras. Aqu, la distancia de Manhattan es de 6.7.
Esto significa que si se midiera la distancia movindose solo a lo largo de las lneas rectas del eje (como moverse por una cuadrcula de calles), la distancia total entre las dos muestras sera 6.7.
Contexto:

La distancia de Manhattan tiende a ser ms grande que la Euclidiana cuando se consideran varias dimensiones, ya que acumula las diferencias en cada dimensin sin elevarlas al cuadrado. Proporciona una visin alternativa de la diferencia, especialmente til en datos con alta dimensionalidad o caractersticas escasas.

### 3. Distancia Coseno: 0.07161964128508791
Interpretacin:

La distancia coseno se mide como 1 menos la similitud del coseno entre los dos vectores. Un valor de 0.0716 indica que el ngulo entre los dos vectores (muestras) es relativamente pequeo, lo que significa que los vectores apuntan en direcciones bastante similares.
En otras palabras, la orientacin (pero no la magnitud) de las caractersticas de las dos flores es muy parecida.
Contexto:

La similitud del coseno es especialmente til cuando queremos saber si dos muestras tienen una "direccin" similar en el espacio de caractersticas, independientemente de cun largas o cortas sean las diferencias en sus caractersticas. Este valor bajo de distancia coseno indica que las muestras estn orientadas en una direccin similar.

### 4. Distancia de Mahalanobis: 2.474107848855276
Interpretacin:

La distancia de Mahalanobis es de aproximadamente 2.47, lo cual toma en cuenta la correlacin entre las variables en el conjunto de datos. Esta mtrica mide la distancia de una muestra al centroide de las muestras (media multivariante), normalizada por la dispersin de los datos.
En otras palabras, una distancia de Mahalanobis de 2.47 indica que las muestras estn a 2.47 desviaciones estndar de diferencia cuando se tiene en cuenta la estructura de correlacin del dataset.
Contexto:

Una distancia de Mahalanobis ms baja sugiere que las muestras estn ms cerca del centroide en el contexto de la distribucin de los datos, mientras que una distancia ms alta indica que estn ms alejadas y pueden considerarse ms "anmalas" o diferentes.

### Conclusin General:

Distancia Euclidiana y Manhattan: Ambas muestran que las dos muestras estn moderadamente separadas en el espacio de caractersticas, pero la distancia Manhattan refleja una ruta ms "realista" a lo largo de cada eje.

Distancia Coseno: Muestra que las dos muestras tienen una orientacin similar, sugiriendo que, aunque pueden tener diferencias en la magnitud de algunas caractersticas, estas diferencias ocurren en la misma direccin.

Distancia de Mahalanobis: Considera la correlacin entre caractersticas y muestra que las dos muestras no son extremadamente diferentes cuando se tienen en cuenta estas correlaciones.



```python
from sklearn.datasets import load_wine
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, cosine, mahalanobis #cityblock es manhathan
import pandas as pd

# Cargar el Wine dataset
wine = load_wine()
X = wine.data  # Caractersticas
y = wine.target  # Etiquetas de clase

# Crear un DataFrame para facilitar el manejo
df = pd.DataFrame(X, columns=wine.feature_names)

# Seleccionar dos muestras del dataset para comparar (por ejemplo, la primera y la segunda)
sample1 = df.iloc[0].values #Va al indice 0 y trae todos los datos
sample2 = df.iloc[100].values #Va al indice 10 y trae todos los datos

# Calcular Distancias
dist_euclidiana = euclidean(sample1, sample2)
dist_manhattan = cityblock(sample1, sample2)
dist_coseno = cosine(sample1, sample2)

# Calcular la Distancia de Mahalanobis (requiere la matriz de covarianza)
cov_matrix = np.cov(X.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)
dist_mahalanobis = mahalanobis(sample1, sample2, inv_cov_matrix)

# Imprimir resultados
print(f"Distancia Euclidiana: {dist_euclidiana}")
print(f"Distancia de Manhattan: {dist_manhattan}")
print(f"Distancia Coseno: {dist_coseno}")
print(f"Distancia de Mahalanobis: {dist_mahalanobis}")
```
