## Agregar o eliminar dimensiones

```python
vector = np.array([1,2,3],ndmin=10)
print(vector)
vector.ndim
```

```python
expand = np.expand_dims(np.array([[1,2,3],[4,5,6]]),axis=0)
print(expand)
expand.ndim
```

```python
expand2 = np.expand_dims(expand,axis=1)
print(expand2)
expand2.ndim
```

```python
print(vector, vector.ndim)
vector_2 = np.squeeze(vector)
print(vector_2, vector_2.ndim)
```
